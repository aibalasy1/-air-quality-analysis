```python
data_co = pd.read_csv("upd_sep_targets_with_lags/dataset_CO_with_lags.csv").drop(columns=['Unnamed: 0'])
data_no = pd.read_csv("upd_sep_targets_with_lags/dataset_NO_with_lags.csv").drop(columns=['Unnamed: 0'])
data_pm10 = pd.read_csv("upd_sep_targets_with_lags/dataset_PM10_with_lags.csv").drop(columns=['Unnamed: 0'])
data_pm25 = pd.read_csv("upd_sep_targets_with_lags/dataset_PM2.5_with_lags.csv").drop(columns=['Unnamed: 0'])
data_so2 = pd.read_csv("upd_sep_targets_with_lags/dataset_SO2_with_lags.csv").drop(columns=['Unnamed: 0'])
data_pm10.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

def plot_feature_importances(df, target, test_size=0.2, random_state=42):
    if target not in df.columns:
        raise ValueError(f"Целевая переменная '{target}' отсутствует в DataFrame.")
    df_sorted = df.sort_values("datetime").reset_index(drop=True)

    # Разделяем признаки и целевую переменную
    X = df_sorted.drop(columns=[target, "datetime"])
    y = df_sorted[target]

    split_idx = int(len(df_sorted) * (1 - test_size))
    X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
    # Обучаем модель
    model = RandomForestRegressor(random_state=random_state)
    model.fit(X_train, y_train)

    # Получаем важность признаков
    feature_importances = pd.DataFrame({
        "Feature": X.columns,
        "Importance": model.feature_importances_
    }).sort_values(by="Importance", ascending=False)

    # Выводим таблицу важности признаков
    print(feature_importances)

    # 📊 Визуализация
    plt.figure(figsize=(12, 6))
    sns.barplot(x=feature_importances["Importance"], y=feature_importances["Feature"], palette="viridis")
    plt.xlabel("Важность признака")
    plt.ylabel("Признак")
    plt.title(f"Feature Importances for {target}")
    plt.show()

# 🎯 Использование:
plot_feature_importances(data_no, "NO")  # для NO
plot_feature_importances(data_co, "CO")  # для CO
plot_feature_importances(data_pm10, "PM10")  # для CO
plot_feature_importances(data_pm25, "PM2.5")  # для CO
plot_feature_importances(data_so2, "SO2")  # для CO
```

```python
from sklearn.linear_model import LassoCV, ElasticNetCV
from sklearn.preprocessing import StandardScaler
import pandas as pd

def select_features_by_model(df, target, method='lasso', alpha_values=np.logspace(-3, 2, 50), cv=5):
    # Удаляем пропуски
    df = df.dropna().copy()
    
    # Выделяем X и y
    X = df.drop(columns=[target, "datetime"])
    y = df[target]

    # Масштабирование признаков
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Выбор модели
    if method == 'lasso':
        model = LassoCV(alphas=alpha_values, cv=cv, random_state=42).fit(X_scaled, y)
    elif method == 'elasticnet':
        model = ElasticNetCV(alphas=alpha_values, l1_ratio=[0.1, 0.5, 0.9], cv=cv, random_state=42).fit(X_scaled, y)
    else:
        raise ValueError("method must be 'lasso' or 'elasticnet'")
    
    # Получение важности признаков
    coef = pd.Series(model.coef_, index=X.columns)
    selected_features = coef[coef != 0].index.tolist()

    print(f"\nSelected features by {method.title()} ({len(selected_features)} out of {len(X.columns)}):")
    print(selected_features)

    # Вернуть датафрейм с отобранными признаками
    selected_df = df[["datetime", target] + selected_features]
    return selected_df
```

```python
# CO
selected_co = select_features_by_model(data_co, target="CO", method="elasticnet")

# NO
selected_no = select_features_by_model(data_no, target="NO", method="elasticnet")

# PM2.5
selected_pm25 = select_features_by_model(data_pm25, target="PM2.5", method="elasticnet")

# PM10
selected_pm10 = select_features_by_model(data_pm10, target="PM10", method="elasticnet")

# SO2
selected_so2 = select_features_by_model(data_so2, target="SO2", method="elasticnet")
```

```python
def remove_outliers(df, target_col, factor=1.5):
    df = df.copy()
    Q1 = df[target_col].quantile(0.25)
    Q3 = df[target_col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor * IQR
    upper = Q3 + factor * IQR
    return df[(df[target_col] >= lower) & (df[target_col] <= upper)]
```

```python
selected_co = remove_outliers(data_co, "CO")
selected_no = remove_outliers(data_no, "NO")
selected_pm10 = remove_outliers(data_pm10, "PM10")
selected_pm25 = remove_outliers(data_pm25, "PM2.5")
selected_so2 = remove_outliers(data_so2, "SO2")
```

```python
import pandas as pd

def drop_columns(*datasets):
    columns_to_drop = [
        'season_Winter', 'precipitation', 'season_Spring', 
        'season_Fall', 'season_Summer', 'is_heating', 'precipitation'
    ]
    
    for dataset in datasets:
        dataset.drop(columns=columns_to_drop, inplace=True, errors='ignore')
    
drop_columns(selected_co, selected_no, selected_pm10, selected_pm25, selected_so2)

data_co.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / np.maximum(y_true, 1e-8))) * 100

def nash_sutcliffe_efficiency(y_true, y_pred):
    return 1 - np.sum((y_true - y_pred)**2) / np.sum((y_true - np.mean(y_true))**2)

def train_and_evaluate_models(df, target, datetime_col="datetime"):
    df = df.dropna()
    X = df.drop(columns=[target, datetime_col])
    y = df[target]
    dates = df[datetime_col]

    # Chronological split
    train_size = int(len(df) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    dates_test = dates[train_size:]

    models = {
        "Linear Regression": LinearRegression(),
        "Ridge": Ridge(alpha=1.0),
        "Lasso": Lasso(alpha=0.1),
        "ElasticNet": ElasticNet(alpha=0.1, l1_ratio=0.5),
        "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
        "XGBoost": xgb.XGBRegressor(n_estimators=100, random_state=42)
    }

    results = {}

    # Actual vs Predicted plots
    plt.figure(figsize=(15, 12))
    for i, (name, model) in enumerate(models.items(), 1):
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Save metrics
        results[name] = {
            "R²": r2_score(y_test, y_pred),
            "MAE": mean_absolute_error(y_test, y_pred),
            "MAPE": mean_absolute_percentage_error(y_test, y_pred),
            "NSE": nash_sutcliffe_efficiency(y_test, y_pred)
        }

        # Plot actual vs predicted
        plt.subplot(3, 2, i)
        plt.plot(dates_test, y_test, label="Actual", color="black")
        plt.plot(dates_test, y_pred, label="Predicted", color="red")
        plt.title(f"{name} — Actual vs Predicted")
        plt.xlabel("Date")
        plt.ylabel(target)
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Print and plot results
    results_df = pd.DataFrame(results).T
    print(results_df)

    # Plot R² and NSE
    results_df[["R²", "NSE"]].plot(kind="bar", figsize=(10, 5), colormap="viridis")
    plt.title("Model Comparison: R² and NSE")
    plt.ylabel("Score")
    plt.xticks(rotation=0)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.ylim(-1, 1.1)
    plt.show()

    # Plot MAE and MAPE
    results_df[["MAE", "MAPE"]].plot(kind="bar", figsize=(10, 5), colormap="plasma")
    plt.title("Model Comparison: MAE and MAPE")
    plt.ylabel("Error")
    plt.xticks(rotation=0)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
```

```python
train_and_evaluate_models(selected_co, "CO")
```
